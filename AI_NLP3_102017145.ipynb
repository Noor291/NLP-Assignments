```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Exercise 1

```python
fig=plt.figure()
```

```python
ax=fig.add_axes([0,0,1,1])
```

```python
t=ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
fig
```

Exercise 2

```python
obj=plt.figure()
```

```python
h=obj.add_axes([0,0,1,1])
i=obj.add_axes([0.2,0.5,0.2,0.2])
obj
```

```python
h.plot(x,y)
i.plot(x,y)
obj
```

```python
h.set_xlabel("x")
h.set_ylabel("y")
i.set_xlabel("x")
i.set_ylabel("y")
obj
```
